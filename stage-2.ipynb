```python
from ultralytics import YOLO
import yaml

# Use fixed dataset path
dataset_yaml_original = "/kaggle/input/yolo-dataset/yolo_dataset/dataset.yaml"
fixed_yaml = "/kaggle/working/fixed_dataset.yaml"

with open(dataset_yaml_original, "r") as f:
    yml = yaml.safe_load(f)

yml['path'] = "/kaggle/input/yolo-dataset/yolo_dataset"

with open(fixed_yaml, "w") as f:
    yaml.safe_dump(yml, f)
```

```python
model = YOLO("/kaggle/input/motor-det-stage2-first60-update/weights/best.pt")

# Start a **new training run**, initialized from this checkpoint
results = model.train(
    data=fixed_yaml,
    epochs=60,  # This is now 60 **additional** epochs from the pretrained state
    batch=1,
    imgsz=1024,
    optimizer='SGD',
    lr0=5e-4,
    lrf=1e-5,
    momentum=0.937,
    weight_decay=0.002,
    patience=30,
    freeze=0,
    mosaic=1.0,
    close_mosaic=30,
    mixup=0.1,
    copy_paste=0.5,
    degrees=90.0,
    translate=0.1,
    scale=0.7,
    shear=10.0,
    flipud=0.5,
    fliplr=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    multi_scale=False,
    cos_lr=True,
    project="/kaggle/working/yolo_weights",
    name="motor_det_stage2_continue",  # give a new run name
    exist_ok=True,
    verbose=True
)
```
